# Wysłanie żądania dostępu do zasoby (strony WWW)

```python
headers = {
    "Cookie": "sv3=1.0_25f9384c-1e05-11f1-baa5-95e071b19cb5",
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36",
    "Host": "www.ceneo.pl"
}
```

```python
url = "https://www.ceneo.pl/124893467#tab=reviews"
response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
page_dom = BeautifulSoup(response.text, 'html.parser')
```

```python
opinions = page_dom.select("div.js_product-review:not(.user-post--highlight)")
print(type(opinions))
print(len(opinions))
```

```python
pierwsza_opinia = page_dom.select_one("div.js_product-review:not(.user-post--highlight)")
print(type(pierwsza_opinia))
```
